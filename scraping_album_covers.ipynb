```python
import os
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import urllib
import re

# Spotify API credentials
CLIENT_ID = 'ae8c8c49346c4c76a620ff7fec0629ea'
CLIENT_SECRET = 'e2a47e33eaa94fa8bdf917652754abf5'

# Initialize Spotipy client
client_credentials_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
output_folder = "/content/drive/MyDrive/AppliedCV_Spring2024/CL Final Project/album_covers"

def download_album_cover(id, song_name, artist_name):
    query = f"{song_name} {artist_name}"
    results = sp.search(q=query, limit=1)
    if results['tracks']['items']:
        track = results['tracks']['items'][0]
        album_cover_url = track['album']['images'][0]['url']
        album_name = track['album']['name']
        image_data = urllib.request.urlopen(album_cover_url).read()
        song_info = "_".join([str(id), song_name, artist_name])
        song_info = song_info.replace("/", "")
        with open(f"{os.path.join(output_folder, song_info)}.jpg", "wb") as f:
            f.write(image_data)
        print(f"Downloaded album cover for '{song_name}' by '{artist_name}'")
    else:
        print(f"No results found for '{song_name}' by '{artist_name}'")
```
