```python
# for data transformation
import numpy as np
# for visualizing the data
import matplotlib.pyplot as plt
# for opening the media file
import scipy.io.wavfile as wavfile
import os
from tqdm import tqdm

```

```python
folder = '../../mnt/disks/songsnap/audio'
all_files = os.listdir(folder)

def generateSpec(file_name, folder):
    if file_name.endswith('.wav'):
        path = os.path.join(folder, file_name)
        try: 
            Fs, aud = wavfile.read(path)

        # Calculate the number of samples for the first 0-10 seconds
            start_sample = 0  # Start at 0 seconds
            end_sample = Fs * 10  # End at 10 seconds
            if aud.ndim == 1:  # Check if audio data is 1-dimensional
                    aud = aud[:, np.newaxis]  # Convert 1D array to 2D
                # Calculate the number of samples for the first 0-10 seconds
            start_sample = 0  # Start at 0 seconds
            end_sample = Fs * 10  # End at 10 seconds
            if aud.shape[0] < end_sample:
                end_sample = aud.shape[0]
            
            # Select the portion of the audio corresponding to the desired duration
            aud = aud[start_sample:end_sample, 0]
            
            # Create the spectrogram
            powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(aud, Fs=Fs)
            
            epsilon = 1e-10
            powerSpectrum[powerSpectrum == 0] = epsilon
       
            Z = 10. * np.log10(powerSpectrum)
            
            new_filename = file_name[:-4] + '.jpg'
            
            save_dir = '../../mnt/disks/songsnap/spec'
            full_path = os.path.join(save_dir, new_filename)
            plt.imsave(full_path, Z, cmap='viridis') 

        except Exception as e:
            print(f"Error processing file {file_name}: {e}")
```

```python
for file_name in tqdm(all_files[0:2000]): 
    generateSpec(file_name, folder)
```
